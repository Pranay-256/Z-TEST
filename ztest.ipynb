```python
import numpy as np
from scipy.stats import norm
```

```python
#alpha = 0.05
alpha = 0.05
sample = [172,174,168,169,171,173,175,170,169,172]
population_mean= 170
population_std = 3
sample_mean=np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(n))
z_score
```

```python
p_value = 2*(1-norm.cdf(abs(z_score))) #cdf = cumalative distribution frequency
```

```python
p_value
```

```python
if p_value<alpha:
    print("I will reject the null hypotheseis")
else:
    print("I wll accept the null hypothesis")
```
